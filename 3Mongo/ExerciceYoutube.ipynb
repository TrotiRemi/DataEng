```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head()
```

## Question 0

### Netoyer les données

```python
len(df_youtube)
```

```python
df_youtube.info()
```

```python
df_youtube["date"] = pd.to_datetime(df_youtube["date"])
```

```python
df_youtube["date"] = pd.to_datetime(df_youtube["date"], format='%d-%m-%Y')
```

```python
df_youtube = df_youtube.dropna()
len(df_youtube)
```

### Importer les données

```python
db_apple = client.apple
collection_apple = db_apple['apple']
```

```python
data_json = df_youtube.to_dict(orient='records')
collection_apple.insert_many(data_json)
```

## Question 1  

```python
apple = collection_apple.find({"channel_title": "Apple"})
for video in apple:
    print(video)
```

## Question 2

```python
unique_categories = df_youtube['category_id'].unique()
print(unique_categories)
```

```python
len(unique_categories)
```

## Question 3

```python
for doc in collection_apple.find():
    # Vérifier si le champ "tags" existe
    if "tags" in doc and isinstance(doc["tags"], str):
        # Découper les tags en liste
        tags_list = doc["tags"].split('|')  # Utiliser le séparateur "|"
        
        # Mettre à jour le document avec les tags découpés
        collection_apple.update_one(
            {"_id": doc["_id"]},  # Filtrer par l'ID du document
            {"$set": {"tags": tags_list}}  # Mettre à jour le champ "tags"
        )

print("Mise à jour des tags terminée.")
```

## Question 4

```python
most_viewed_videos = collection_apple.find().sort("views", -1).limit(5)
for video in most_viewed_videos:
    print(video)
```

## Question 5

```python
mean_views_by_category = df_youtube.groupby("category_id")["views"].mean()
mean_views_by_category
```

## Question 6 

```python
mean_likes_by_channel = df_youtube.groupby("channel_title")["likes"].mean()

top_channels = mean_likes_by_channel.sort_values(ascending=False).head(5)
```
